### 5.1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

#### **Hypothesis**
* $H_0: \mu_1 = \mu_2$ (There is NO difference in median value of houses bounded by the Charles river)
* $H_1: \mu_1 \neq \mu_2$ (There is a difference in median value of houses bounded by the Charles river)

```python
import scipy.stats as st
```

```python
st.levene(boston_df[boston_df['CHAS'] == 0]['MEDV'],
          boston_df[boston_df['CHAS'] == 1]['MEDV'],
          center='mean')
```

Since the p-value < 0.05 we can not assume equality of variance

```python
st.ttest_ind(boston_df[boston_df['CHAS'] == 0]['MEDV'],
             boston_df[boston_df['CHAS'] == 1]['MEDV'],
             equal_var=False)
```

#### **Conclusion**
Since the p-value < 0.05 we reject the null hypothesis $H_0$ as there is enough proof that there is a statistical difference in Median value of owner-occupied homes based on Charles river variable

### 5.2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

#### **Hypothesis**
* $H_0: \mu_1 = \mu_2 = \mu_3$ (All 3 means are equal)
* $H_1:$ At least one of the means differ

```python
st.levene(boston_df[boston_df['age_group'] == '35 years and younger']['MEDV'],
          boston_df[boston_df['age_group'] == 'Between 35 and 70 years']['MEDV'],
          boston_df[boston_df['age_group'] == '70 years and older']['MEDV'],
          center='mean')
```

Since the p-value > 0.05, the variance is equal.
Now we separate the 3 samples (one for each age category) into a variable each

```python
younger35 = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
from35to70 = boston_df[boston_df['age_group'] == 'Between 35 and 70 years']['MEDV']
older70 = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
```

```python
# One-way ANOVA
f_stat, p_value = st.f_oneway(younger35, from35to70, older70)
print(f'F_statistics = {f_stat}\np-value = {p_value}')
# print("F_statistics: {0}, P-value: {1}".format(f_stat, p_value))
```

#### **Conclusion**
Since the p-value < 0.05, we reject the null hypothesis $H_0$ as there is a significant evidence that at least one of the means differ.

### 5.3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

#### **Hypothesis**
* $H_0:$ Both do NOT correlate
* $H_1:$ Both correlate

```python
st.pearsonr(x=boston_df['NOX'], y=boston_df['INDUS'])
```

#### **Conclusion**
Since the p-value < 0.05, we reject the null hypothesis $H_0$ and conclude that there exists a relationship between NOX and INDUS.

### 5.4. What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

#### **Hypothesis**
* $H_0:$ Distance is NOT correlated with MEDV
* $H_1:$ Distance is correlated with MEDV

```python
import statsmodels.api as sm


X = sm.add_constant(boston_df['DIS'])
y = boston_df['MEDV']

model = sm.OLS(y, X).fit()
predict = model.predict(X)

model.summary()
```

#### **Conclusion**
Since the p-value < 0.05, we reject the null hypothesis $H_0$ and conclude that there exists a relationship between DIS and MEDV.
